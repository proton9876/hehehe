```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python

# Load the iris dataset
iris = load_iris()
X = iris.data
```

```python

# Apply hierarchical clustering algorithm
linkage_matrix = linkage(X, 'ward')  
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python

agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clusters = agg_clustering.fit_predict(X)
```

```python

# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[agg_clusters == 0, 0], X[agg_clusters == 0, 1], s=10, c='red', label='Cluster 1')
plt.scatter(X[agg_clusters == 1, 0], X[agg_clusters == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[agg_clusters == 2, 0], X[agg_clusters == 2, 1], s=50, c='green', label='Cluster 3')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
